Working with Video

```python
import matplotlib.pyplot as plt
import cv2

# Path to your video file
video_path = '/content/drive/MyDrive/Colab Notebooks/MV/Lab 1/test.mp4'

# Initialize video capture object
capture = cv2.VideoCapture(video_path)

# Check if capture object is opened successfully
if not capture.isOpened():
    print("Error: Unable to open video file.")
else:
    while True:
        ret, frame = capture.read()
        if not ret:
            break

        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        plt.imshow(frame_rgb)
        plt.title('Video Frame')  # Optional: Set title for the plot
        plt.axis('off')  # Optional: Turn off axis labels
        plt.show()

        cv2.waitKey(10000)  # Wait for milliseconds
capture.release()
```

Different Model of Images

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
image_lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
image_ycrcb = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)
image_xyz = cv2.cvtColor(image, cv2.COLOR_BGR2XYZ)

# Edge detection using Canny
edges = cv2.Canny(image_gray, 100, 200)

# Display the images
plt.figure(figsize=(25, 10))

# Original Image (BGR)
plt.subplot(3, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image (BGR)')
plt.axis('off')

# RGB Image
plt.subplot(3, 3, 2)
plt.imshow(image_rgb)
plt.title('RGB Image')
plt.axis('off')

# Grayscale Image
plt.subplot(3, 3, 3)
plt.imshow(image_gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

# HSV Image
plt.subplot(3, 3, 4)
plt.imshow(cv2.cvtColor(image_hsv, cv2.COLOR_HSV2RGB))
plt.title('HSV Image')
plt.axis('off')

# LAB Image
plt.subplot(3, 3, 5)
plt.imshow(cv2.cvtColor(image_lab, cv2.COLOR_LAB2RGB))
plt.title('LAB Image')
plt.axis('off')

# YCrCb Image
plt.subplot(3, 3, 6)
plt.imshow(cv2.cvtColor(image_ycrcb, cv2.COLOR_YCrCb2RGB))
plt.title('YCrCb Image')
plt.axis('off')

# XYZ Image
plt.subplot(3, 3, 7)
plt.imshow(cv2.cvtColor(image_xyz, cv2.COLOR_XYZ2RGB))
plt.title('XYZ Image')
plt.axis('off')

# Canny Edge Detection
plt.subplot(3, 3, 8)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/MV/Lab 1/download.jpeg')

# Convert the image to different color models
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
image_lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
image_ycrcb = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)
image_xyz = cv2.cvtColor(image, cv2.COLOR_BGR2XYZ)

# Display the images
plt.figure(figsize=(15, 10))

# Original Image (BGR)
plt.subplot(3, 3, 1)
plt.imshow(image)
plt.title('Original Image (BGR)')
plt.axis('off')

# RGB Image
plt.subplot(3, 3, 2)
plt.imshow(image_rgb)
plt.title('RGB Image')
plt.axis('off')

# Grayscale Image
plt.subplot(3, 3, 3)
plt.imshow(image_gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

# HSV Image
plt.subplot(3, 3, 4)
plt.imshow(image_hsv, cmap='hsv')
plt.title('HSV Image')
plt.axis('off')

# XYZ Image
plt.subplot(3, 3, 5)
plt.imshow(image_xyz, cmap='gray')
plt.title('XYZ Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
